```python
import pandas as pd
import os
import plotly.express as px
```

```python
os.listdir("results")
```

```python
ACR_per_episode = {}
for dir_path in os.listdir("results"):
    df = pd.read_csv(f"results/{dir_path}")
    print(f"ACR {dir_path}: {df.groupby('episode').reward.mean()}")
    ACR_per_episode[dir_path.split('.')[0]] = df.groupby('episode').reward.mean()
```

```python
ACR_per_episode = pd.DataFrame(ACR_per_episode)
print(ACR_per_episode.mean(axis=0))
px.line(ACR_per_episode)
```
